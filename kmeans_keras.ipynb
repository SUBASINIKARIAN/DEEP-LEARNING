```python
import numpy as np
from sklearn.cluster import KMeans
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense

```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load and preprocess the data
data = pd.read_csv("D:\Datasets\Heart.csv")
data["ChestPain"] = data["ChestPain"].astype('category')
data.dtypes
data["ChestPain"]= data["ChestPain"].cat.codes
data.head()

data["Thal"] = data["Thal"].astype('category')
data.dtypes
data["Thal"]= data["Thal"].cat.codes
data.head()

data["AHD"] = data["AHD"].astype('category')
data.dtypes
data["AHD"]= data["AHD"].cat.codes
data.head()


X = data.iloc[:,5:9].values
y = data.iloc[:, -1].values
scaler = StandardScaler()
X = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.16)
```

```python
print(y_train)
```

```python
print(y_test)
```

```python
class RBF():
    def __init__(self, x_train,y_train,n_centers):
        kmeans = KMeans(n_clusters=n_centers)
        kmeans.fit(X_train)
        centers = kmeans.cluster_centers_

        model = keras.Sequential([
            Dense(n_centers, input_shape=(4,), activation='relu', use_bias=False),
            Dense(1, activation='linear', use_bias=False)
        ])

        model.layers[0].set_weights([centers.T])

        model.compile(loss='mse', optimizer='adam',metrics=['accuracy'])

        model.fit(X_train, y_train, epochs=100, batch_size=32)

        accuracy,mse= model.evaluate(X_test, y_test)
        print(" Mean Squared Error & Accuracy:", accuracy,mse)
```

```python
rbf=RBF(X_train,y_train,10)
```
